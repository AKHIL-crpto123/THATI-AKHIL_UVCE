```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = r"C:\Users\akhil\OneDrive\Desktop\Customers.csv"
df = pd.read_csv(file_path)

# Display the first few rows of the data
print("First few records of the data:")
print(df.head())
```

```python
# Check for missing values in each column
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
# Summary statistics for numerical columns
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Data types and info
print("\nData Types:")
print(df.dtypes)
```

```python
# Check unique values in the 'Region' column
print("\nUnique Regions:")
print(df['Region'].unique())
```

```python
# Convert SignupDate to datetime format (if available)
if 'SignupDate' in df.columns:
    df['SignupDate'] = pd.to_datetime(df['SignupDate'], errors='coerce')

# Check for any invalid SignupDate conversion
print("\nMissing or invalid SignupDate entries:")
print(df['SignupDate'].isnull().sum())
```

```python
# Visualizing the customer distribution by region with corrected hue usage
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Region', hue='Region', palette='Set2', legend=False)
plt.title('Distribution of Customers by Region')
plt.show()
```

```python
# Visualize the signup trends over time (if SignupDate exists)
if 'SignupDate' in df.columns:
    plt.figure(figsize=(10, 6))
    df['SignupDate'].hist(bins=20, color='skyblue')
    plt.title('Distribution of Customer Signups Over Time')
    plt.xlabel('Signup Date')
    plt.ylabel('Number of Signups')
    plt.show()
else:
    print("\nSignupDate column not found in the dataset.")
```
